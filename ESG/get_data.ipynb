```python
#Get data from the SIX API
import pandas as pd

# Load the dataset
file_path = 'EUESGMANUFACTURER.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Reshape the data
reshaped_data = data.copy().pivot_table(index='ISIN', columns='ESGFactor', values='ESGFactorAmountLastYear', aggfunc='first')
reshaped_data_units = data.copy().pivot_table(index='ISIN', columns='ESGFactor', values='ESGFactorAmountLastYearUnit', aggfunc='first')

# Display the reshaped data's first few rows
display(reshaped_data.head())
reshaped_data_units.sort_index().head()

print(f'Are all ISIN unique: {len(reshaped_data.index.unique()) == reshaped_data.shape[0]}; Shape: {reshaped_data.shape}')
```

```python
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]
```

```python
df=data.copy()
print(f'{len(df["ISIN"].unique())} companies in the dataset')
print(f'{len(reshaped_data["30020_GHG_Emissions_Scope_1_Value"].unique())-1} scope 1 entries')
print(f'{len(reshaped_data["30060_GHG_Emissions_Scope_2_Value"].unique())-1} scope 2 entries')
print(f'{len(reshaped_data["30100_GHG_Emissions_Scope_3_Value"].unique())-1} scope 3 entries')

#Get entries for which all 3 scope entries are available
scope1 = reshaped_data['30020_GHG_Emissions_Scope_1_Value']
scope2 = reshaped_data['30060_GHG_Emissions_Scope_2_Value']
scope3 = reshaped_data['30100_GHG_Emissions_Scope_3_Value']

#Join the scopes by ISIN and drop zeros, they are in tCO2e/M€ invested
scope = pd.concat([scope1,scope2,scope3], axis=1).dropna()

scope["total"] = scope.sum(axis=1)
scope.drop(scope.loc[scope['total']==0].index, inplace=True)
scope["z-score"] = (scope["total"] - scope["total"].mean())/scope["total"].std(ddof=0)
scope["savings_vs_vanguard"] = (52814 - scope["total"])
scope=scope.sort_values(by="total",ascending=True)
scope.head()

display(scope)
scope.value_counts()
```
